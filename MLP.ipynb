# 多层感知机的理论实现基础：激活函数

如果没有激活函数，多层线性回归实际上等价于单层线性回归

```python
x = torch.arange(-8, 8, 0.1, requires_grad=True)
y = torch.relu(x)
# plt.plot(x.detach(), y.detach()) #, 'x', 'relu(x)')
```

```python
x_value = x.detach()
y_value = y.detach()
plt.plot(x_value, y_value)
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
```

```python
plt.plot(x.detach(), x.grad)
```
